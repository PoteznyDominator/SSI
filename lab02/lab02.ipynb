```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots()

plt.scatter([-1], [1], color="blue")
plt.scatter([1], [1], color="blue")
plt.scatter([0], [0], color="blue")
circle = plt.Circle((0,0), 2, color="red", fill=False)
ax.add_patch(circle)
x = np.linspace(-1, 1, 1000)
ax.plot(x, np.sin(x- np.pi/2), color="yellow")

plt.grid()
plt.xlim([-2,2])
plt.ylim([-3,3])
plt.show()
```

```python
df_attr = pd.read_csv("../lab01/iris-type.txt", sep='\t', header=None)
df = pd.read_csv("../lab01/iris.txt", sep='\t', header=None, names=df_attr[0])
```

```python
fig, axs = plt.subplots(2,2)
fig.tight_layout()

sns.scatterplot(x=df['petal_length_in_cm'], y=df['petal_width_in_cm'], hue=df['class(1=Setosa,2=Versicolour,3=Virginica)'], palette='bright', ax=axs[0,0]).legend([],[], frameon=False)
sns.scatterplot(x=df['sepal_width_in_cm'], y=df['petal_width_in_cm'], hue=df['class(1=Setosa,2=Versicolour,3=Virginica)'], palette='bright', ax=axs[0,1]).legend([],[], frameon=False)
sns.scatterplot(x=df['sepal_length_in_cm'], y=df['petal_width_in_cm'], hue=df['class(1=Setosa,2=Versicolour,3=Virginica)'], palette='bright', ax=axs[1,0]).legend([],[], frameon=False)
sns.scatterplot(x=df['sepal_width_in_cm'], y=df['petal_length_in_cm'], hue=df['class(1=Setosa,2=Versicolour,3=Virginica)'], palette='bright', ax=axs[1,1]).legend([],[], frameon=False)

fig.legend(['Virginica','Setosa','Versicolour'],bbox_to_anchor=(1.2,1))
plt.show()
```
