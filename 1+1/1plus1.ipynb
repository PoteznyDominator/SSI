```python
import random
import numpy as np
import matplotlib.pyplot as plt

def get_func(x):
    return np.sin(x/10.0) * np.sin(x/200)


def one_plus_one(rozrzut, wsp_przyrostu, l_iteracji, zakres_zmiennosci):
    x = random.uniform(zakres_zmiennosci[0], zakres_zmiennosci[1])
    y = get_func(x)
    t = np.linspace(0, 100, 1000)
    
    for i in range(l_iteracji):
        x_pot = x + random.uniform(-rozrzut, rozrzut)
        
        # sprawdzenie zakresu zmiennosci
        x_pot = max(min(x_pot, zakres_zmiennosci[1]), zakres_zmiennosci[0])
        y_pot = get_func(x_pot)
        
        if y_pot >= y:
            x = x_pot
            y = y_pot
            rozrzut *= wsp_przyrostu
            
            plt.plot(t, get_func(t), color='blue')
            plt.scatter(x, y, color='red', s=100)
            plt.title(f'Numer iteracji: {i}, x: {x}, y: {y}, rozrzut: {rozrzut}')
            plt.show()
        else:
            rozrzut /= wsp_przyrostu
        
        # print(f'Numer iteracji: {i}, x: {x}, y: {y}, rozrzut: {rozrzut}')
    
    plt.plot(t, get_func(t), color='blue')
    plt.scatter(x, y, color='red', s=100)
    plt.title("Finałowy wykres")
    plt.show()

         
        
one_plus_one(10, 1.1, 100, (0,100))        
```
